```python
#currentttttttttt

#importing the library 
#for playing and loading mp3 files 
#file dialog
#message box

from tkinter import *
from pygame import mixer
from tkinter import filedialog
import tkinter.messagebox

#instance
window = Toplevel()

mixer.init()


#for the popup geometry
window.geometry('500x600')

#title of the window
window.title('music player')

#open the file dialog and load
def browse_file():
    global filename
    filename = filedialog.askopenfilename()
    
def help_menu():
    tkinter.messagebox.showinfo("Okkk", "How can i help You")
    
#menu class
menubar = Menu(window)
submenu = Menu(menubar, tearoff=0)
window.config(menu=menubar)


menubar.add_cascade(label = 'File', menu=submenu)
submenu.add_command(label = 'Open', command = browse_file)
submenu.add_command(label = "Exit", command = window.destroy)

submenu = Menu(menubar, tearoff=0)
menubar.add_cascade(label = 'About Us', menu = submenu)
submenu.add_command(label = 'Help', command = help_menu)

#instance for label class
textlabel =Label(window, text="Play Button ")
textlabel.grid()



def play_music():
    try:
        mixer.music.load(filename)
        mixer.music.play()
        statusbar['text'] = "Music is Being Played"
    except:
        tkinter.messagebox.showerror("Error", "File Not Found")
    
                
def stop_music():
    mixer.music.stop()
    statusbar['text'] = "Music is Stopped"
    
def set_volume(value):
    volume = int(value)/100 #convert string into int
    mixer.music.set_volume(volume) #range 0 to 1

def rewind_music():
    play_music()
    statusbar['text'] = "Music is Rewinded"

    
#frame for buttons
frame = Frame(window)
frame.grid(pady=10, padx=10, row=1)


#the image of the button & button
photo = PhotoImage(file='play2.png')
playbutton = Button(frame, image= photo , command=play_music)
playbutton.grid(row=0, column=0 ,padx=10)

#stopbutton
stop_photo = PhotoImage(file='pause.png')
stopbutton = Button(frame, image = stop_photo, command=stop_music)
stopbutton.grid(row=0 ,column=1, pady=10, padx=10)

#stoplbutton label
stoplabel  = Label(window, text="Stop Button")
stoplabel.grid()

bottomframe = Frame(window)
bottomframe.grid()

rewindphoto = PhotoImage(file='rewind.png')
rewindbutton = Button(window, image=rewindphoto, 
                      command=rewind_music)

rewindbutton.grid(row=0, column=0)


#scale for volume
scale = Scale(bottomframe, from_=0, to=100, orient = HORIZONTAL, command = set_volume)
scale.set(70)
scale.grid(row=0, column=1, pady=15)

#statusbar shows

statusbar = Label(window, text="Keep enjoying the music", 
                  relief=SUNKEN,anchor=W)

statusbar.grid()


#each updation done by running it
window.mainloop()
```
